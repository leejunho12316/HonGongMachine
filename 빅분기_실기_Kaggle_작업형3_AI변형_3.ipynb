```python
import pandas as pd
import numpy as np

# 시드 고정
np.random.seed(1234)

# 데이터셋 1 생성 (합격 여부 예측)
data1 = {
    'study_hours': np.random.randint(1, 15, 500),
    'attendance': np.random.randint(50, 100, 500),
    'pass_yn': np.random.randint(0, 2, 500)
}
df1 = pd.DataFrame(data1)

train = df1.iloc[:400]
test = df1.iloc[400:]

from statsmodels.formula.api import logit
model = logit('pass_yn~study_hours', data=train).fit()

res = np.exp(model.params['study_hours'])
print(round(res, 4))
```

```python
import pandas as pd
import numpy as np
from statsmodels.formula.api import logit
from sklearn.metrics import accuracy_score

# =====================================================
# [문제 1 풀이] 오즈비 계산
# =====================================================
# 1. 데이터 분할
train1 = df1.iloc[:400]
test1 = df1.iloc[400:]

# 2. 모델 학습 (study_hours -> pass)
# 공식: logit('종속변수 ~ 독립변수', data=데이터)
model1 = logit('pass_yn ~ study_hours', data=train1).fit(disp=0)
# disp=0은 최적화 과정 로그 출력을 억제함(필수는 아님)

# 3. 오즈비 계산 (np.exp(계수))
odds_ratio = np.exp(model1.params['study_hours'])

print(f"문제 1 정답: {round(odds_ratio, 4)}")
```

```python
# 데이터셋 2 생성 (스팸 메일 분류)
data2 = {
    'word_count': np.random.normal(100, 20, 400),
    'link_count': np.random.randint(0, 10, 400),
    'is_spam': np.random.choice([0, 1], 400, p=[0.7, 0.3])
}
df2 = pd.DataFrame(data2)

train = df2.iloc[:300]
test = df2.iloc[300:]

from statsmodels.formula.api import logit
model = logit('is_spam~link_count', data=train).fit()

df2['pred'] = model.predict(test)
df2['pred'] = (df2['pred'] > 0.5).astype(int)
df2['score'] = (df2['is_spam'] == df2['pred']).astype(int)
print(round(df2['score'].mean(),3))
```

```python

# =====================================================
# [문제 2 풀이] 정확도 계산
# =====================================================
# 1. 데이터 분할
train2 = df2.iloc[:300]
test2 = df2.iloc[300:]

# 2. 모델 학습 (link_count -> is_spam)
model2 = logit('is_spam ~ link_count', data=train2).fit(disp=0)

# 3. 예측 (predict)
pred_prob = model2.predict(test2)

# 4. 클래스 분류 (0.5 초과 1, 이하 0)
pred_class = (pred_prob > 0.5).astype(int)

# 5. 정확도 계산
accuracy = accuracy_score(test2['is_spam'], pred_class)

print(f"문제 2 정답: {round(accuracy, 3)}")
```

```python
# 데이터셋 3 생성 (대출 연체 여부)
np.random.seed(1234)
data3 = {
    'income': np.random.normal(5000, 1000, 300),
    'debt': np.random.normal(2000, 500, 300),
    'default': np.random.choice([0, 1], 300, p=[0.8, 0.2])
}
df3 = pd.DataFrame(data3)

train = df3.iloc[:200]
test = df3.iloc[200:]

from statsmodels.formula.api import logit
model = logit('default~debt',data = train).fit()

print(round(model.llf + model.pvalues['debt'], 2))
```

```python
# =====================================================
# [문제 3 풀이] 로짓 우도 및 P-value
# =====================================================
# 1. 데이터 분할
train3 = df3.iloc[:200]
test3 = df3.iloc[200:]

# 2. 모델 학습 (debt -> default)
model3 = logit('default ~ debt', data=train3).fit(disp=0)

# 3. 로짓 우도 (Log-Likelihood)
log_likelihood = model3.llf

# 4. 유의확률 (P-value)
p_val = model3.pvalues['debt']

# 5. 두 값의 합 계산
result3 = log_likelihood + p_val

print(f"문제 3 로짓 우도: {log_likelihood}")
print(f"문제 3 P-value: {p_val}")
print(f"문제 3 정답(합): {round(result3, 2)}")
```

```python
import pandas as pd
import numpy as np

# 시드 고정 (재현성을 위해 필수)
np.random.seed(2024)

# 데이터셋 1: 질병 진단 (diabetes)
data1 = {
    'age': np.random.randint(20, 80, 500),
    'bmi': np.random.normal(25, 5, 500),
    'bp': np.random.randint(60, 140, 500),
    'diabetes': np.random.choice([0, 1], 500, p=[0.7, 0.3])
}
df1 = pd.DataFrame(data1)

# 데이터셋 2: 통신사 해지 방어 (churn)
data2 = {
    'usage_minutes': np.random.normal(300, 100, 500),
    'customer_service_calls': np.random.randint(0, 10, 500),
    'contract_period': np.random.randint(1, 36, 500),
    'churn': np.random.choice([0, 1], 500, p=[0.8, 0.2])
}
df2 = pd.DataFrame(data2)

# 데이터셋 3: 공정 불량 예측 (defect)
data3 = {
    'temperature': np.random.normal(200, 20, 500),
    'pressure': np.random.normal(50, 10, 500),
    'humidity': np.random.randint(30, 70, 500),
    'defect': np.random.choice([0, 1], 500, p=[0.9, 0.1])
}
df3 = pd.DataFrame(data3)

print("데이터 생성 완료. 문제 풀이를 시작하게.")
```

```python
import pandas as pd
import numpy as np
from statsmodels.formula.api import logit

# =====================================================
# [문제 1 풀이]
# =====================================================
# 1. 데이터 분할
train1 = df1.iloc[:350]
test1 = df1.iloc[350:]

# 2. 모델 학습 (logit 사용)
# logit 함수는 잔차이탈도 속성(.deviance)이 바로 없으므로 llf를 이용해 계산한다.
model1 = logit('diabetes ~ age + bmi', data=train1).fit(disp=0)

# 3. 잔차이탈도 계산: -2 * Log-Likelihood
deviance1 = -2 * model1.llf

print(f"문제 1 정답: {round(deviance1, 4)}")


# =====================================================
# [문제 2 풀이]
# =====================================================
# 1. 데이터 분할
train2 = df2.iloc[:400]
test2 = df2.iloc[400:]

# 2. 모델 학습
model2 = logit('churn ~ usage_minutes + customer_service_calls', data=train2).fit(disp=0)

# 3. 잔차이탈도 계산
deviance2 = -2 * model2.llf

print(f"문제 2 정답: {round(deviance2, 3)}")


# =====================================================
# [문제 3 풀이] GLM 방식 활용 (참고용)
# =====================================================
from statsmodels.formula.api import glm
import statsmodels.api as sm

# 1. 데이터 분할
train3 = df3.iloc[:300]
test3 = df3.iloc[300:]

# 2. 모델 학습 (GLM 사용)
# GLM을 사용하고 family=sm.families.Binomial()을 지정하면 로지스틱 회귀와 같다.
# GLM 결과 객체는 .deviance 속성을 직접 가지고 있다.
model3 = glm('defect ~ temperature + pressure + humidity',
             data=train3,
             family=sm.families.Binomial()).fit()

# 3. 잔차이탈도 출력
# model3.deviance 는 정확히 -2 * model3.llf 와 같다.
print(f"문제 3 정답: {round(model3.deviance, 2)}")

# (검증) logit으로 했을 때와 같은지 확인
model3_logit = logit('defect ~ temperature + pressure + humidity', data=train3).fit(disp=0)
print(f"(검증) Logit 이용값: {round(-2 * model3_logit.llf, 2)}")
```

```python
import pandas as pd
import numpy as np

# 시드 고정
np.random.seed(2024)

# 데이터셋 1: 에너지바 단백질 함량 (목표: 20g)
# 실제로는 20보다 조금 더 들어있는 데이터
data1 = {
    'protein': np.random.normal(21.5, 2, 30).tolist()
}
df1 = pd.DataFrame(data1)

# 데이터셋 2: 고객 대기 시간 (목표: 10분)
# 실제로는 10분보다 적게 걸리는 데이터 (일부는 정확히 10분인 경우 포함)
data2 = {
    'wait_time': np.random.normal(8.5, 1.5, 40).tolist()
}
# 0 처리 경고 확인용 데이터 추가 (정확히 10분인 값)
df2 = pd.DataFrame(data2)
df2.loc[0, 'wait_time'] = 10.0
df2.loc[1, 'wait_time'] = 10.0

# 데이터셋 3: 볼트 길이 (목표: 50mm)
# 50mm와 차이가 있는지 검정
data3 = {
    'length': np.random.normal(50.5, 1.0, 50).tolist()
}
df3 = pd.DataFrame(data3)

print("데이터 생성 완료")
```

```python
import pandas as pd
from scipy.stats import wilcoxon

# =====================================================
# [문제 1 풀이] 대립가설: 중앙값 > 20 (greater)
# =====================================================
mu1 = 20
# 귀무가설: median == 20
# 대립가설: median > 20
# 데이터에서 20을 뺀 값이 0보다 큰지 검정하는 것과 같음
stat1, p1 = wilcoxon(df1['protein'] - mu1, alternative='greater')

print(f"문제 1 정답 (p-value): {round(p1, 4)}")


# =====================================================
# [문제 2 풀이] 대립가설: 중앙값 < 10 (less)
# =====================================================
mu2 = 10
# 귀무가설: median == 10
# 대립가설: median < 10
# 데이터에 정확히 10인 값이 있으면 차이가 0이 되어 경고 메시지가 뜰 수 있음 (정상임)
stat2, p2 = wilcoxon(df2['wait_time'] - mu2, alternative='less')

print(f"문제 2 정답 (p-value): {round(p2, 3)}")


# =====================================================
# [문제 3 풀이] 대립가설: 중앙값 != 50 (two-sided)
# =====================================================
mu3 = 50
# 귀무가설: median == 50
# 대립가설: median != 50 (기본값이 two-sided이므로 생략 가능하지만 명시하는 게 좋음)
stat3, p3 = wilcoxon(df3['length'] - mu3, alternative='two-sided')

print(f"문제 3 정답 (통계량): {int(stat3)}")
```
