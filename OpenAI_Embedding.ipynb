#OpenAI Embedding Model 준비

```python
from openai import OpenAI
client = OpenAI(
    api_key = ""
)
```

```python
def get_embedding(text):
  embedding = client.embeddings.create(
      input= text,
      model = 'text-embedding-ada-002'
  )

  return embedding.data[0].embedding
```

```python
result = get_embedding('안녕하세요')
print(result)
```

#데이터 준비

```python
import pandas as pd
import numpy as np

data = [
    '학살의 현장에서 난 피어오른다. 붉은 여명에 피어나는 꽃처럼.',  # 진 (예술/광기)
    '정확성. 그게 바로 도살자와 외과 의사의 차이지.',             # 카밀 (정교함/오만함)
    '왕관을 쓴 자들은 모두, 내 사슬 맛을 보게 될 거야.',           # 사일러스 (반역/혁명)
    '죽음은 바람과 같지. 늘 내 곁에 있으니.',                     # 야스오 (고독/바람)
    '보이지 않는 검이 가장 무서운 법.',                           # 제드 (암살/그림자)
]
df = pd.DataFrame(data, columns = ['text'])
df
```

#데이터 임베딩

```python
df['embeddings'] = df['text'].apply(lambda x : get_embedding(x))
df
```

```python
import numpy as np

def cos_sim(A,B):
  return np.dot(A,B)/(np.linalg.norm(A) * np.linalg.norm(B))

def return_answer_candidate(df, query):
  query_embedding = get_embedding(query)

  df['similarity'] = df['embeddings'].apply(lambda x : cos_sim(x, query_embedding))
  #df['similarity'] = df['embeddings'].apply(lambda x : cos_sim(np.array(x), np.array(query_embedding)))

  df.sort_values(by=['similarity'], ascending=False, inplace=True)

  return df.head(3)
```

```python
return_answer_candidate(df, "외과 의사")
```
