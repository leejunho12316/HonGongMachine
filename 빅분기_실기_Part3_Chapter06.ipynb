#01 단일 표본 검정

```python
import pandas as pd
df = pd.DataFrame({
    'Caffeine(mg)' : [
        94.2, 93.7, 95.5, 93.9, 94.0, 95.2, 94.7, 93.5, 92.8, 94.4,
        93.8, 94.6, 93.3, 95.1, 94.3, 94.9, 93.9, 94.8, 95.0, 94.2,
        93.7, 94.4, 95.1, 94.0, 93.6
    ]
})
```

```python
#1. 표본 데이터의 평균
m = df['Caffeine(mg)'].mean()
print(m)
```

```python
#2. Shapiro-Wilk p-value
from scipy import stats
t_statistics, p_value = stats.shapiro(df['Caffeine(mg)'])
print(p_value) #정규성을 만족한다
```

```python
#3. 단일 표본 t-검정의 검정 통계량
t_statistics, p_value = stats.ttest_1samp(df['Caffeine(mg)'], 95, alternative='less')
print(t_statistics)
print(p_value) #format 어캐쓰는거였지
```

```python
print('기각')
```

#02 독립 표본 검정

```python
import pandas as pd
df = pd.DataFrame({
    '충전기' : ['New'] * 10 + ['Old'] * 10,
    '충전시간' : [
        1.5, 1.6, 1.4, 1.7, 1.5, 1.6, 1.7, 1.4, 1.6, 1.5,
        1.7, 1.8, 1.7, 1.9, 1.8, 1.7, 1.8, 1.9, 1.7, 1.6
    ]
})
```

```python
df_new = df[df['충전기'] == 'New']['충전시간']
df_old = df[df['충전기'] == 'Old']['충전시간']

#위의 가설을 검정하기 위한 검정통계량
from scipy import stats
t_statistics, p_value = stats.ttest_ind(df_new, df_old, alternative='less', equal_var = True)
print(t_statistics)


#위의 통계량에 대한 p-value
print(p_value)

#0.05유의수준에서 귀무가설 채택/기각?
print('기각')
```

#03 대응 표본 검정

```python
import pandas as pd
df = pd.DataFrame({
    'User' : list(range(1,11)),
    '기존방법' : [60.4, 60.7, 60.5, 60.3, 60.8, 60.6, 60.2, 60.5, 60.7, 60.4],
    '새로운방법' : [59.8, 60.2, 60.1, 59.9, 59.7, 58.4, 57.0, 60.3, 59.6, 59.8]
})
```

```python
#md의 표본 평균 (새로운 - 기존)
df['차이'] = df['새로운방법']-df['기존방법']
print(df['차이'].mean())

#검정통계량
from scipy import stats
t, p = stats.ttest_rel(df['새로운방법'], df['기존방법'], alternative='less')
print(t)

#p_value
print(p)

#채택 기각
print('기각')
```

#04 일원 분산 분석

```python
import pandas as pd
df = pd.read_csv('math.csv')
df.sample(5)
```

```python
#각 그룹 수학 성적 Shapiro-Wilk 검정의 p-value
df_A = df[df['groups'] == 'group_A']['scores']
df_B = df[df['groups'] == 'group_B']['scores']
df_C = df[df['groups'] == 'group_C']['scores']
df_D = df[df['groups'] == 'group_D']['scores']

from scipy import stats
_, p_A = stats.shapiro(df_A)
_, p_B = stats.shapiro(df_B)
_, p_C = stats.shapiro(df_C)
_, p_D = stats.shapiro(df_D)
print(p_A, p_B, p_C, p_D)

#4개 그룹의 등분산성 Levene 검정의 p-value
_, p = stats.levene(df_A, df_B, df_C, df_D)
print(p)

#유의수준 0.05 귀무가설 채택/기각?
print('채택')

#그룹변수의 자유도
from statsmodels.formula.api import ols
import statsmodels.api as sm
model = ols('scores~groups', data =df).fit()
res = sm.stats.anova_lm(model)

print(res.loc['groups','df'])

#잔차의 자유도
print(res.loc['Residual','df'])

#성적의 제곱합
print(res.loc['groups','sum_sq'])

#성적의 평균 제곱
print(res.loc['groups','mean_sq'])

#F-통계량 값
print(res.loc['groups', 'F'])

#성적에 대한 p-value
print(res.loc['groups', 'PR(>F)'])
```

#06 적합도 검정

```python
import pandas as pd
df = pd.DataFrame({
    '경험 수':[550, 250, 100, 70, 30],
    '전국적인 경향(%)' : [1000*0.6, 1000*0.25, 1000*0.08, 1000*0.05, 1000*0.02]
})
df
```

```python
#이도시의 교통사고 경험자 중 5회 이상의 비율
print(30/1000)

#이도시의 교통사고 분포가 전국적인 경향을 띠는지
from scipy import stats
s,p = stats.chisquare(df['경험 수'], df['전국적인 경향(%)'])
print(s)
print(p)
print('기각')
```

```python
import pandas as pd
df = pd.DataFrame({
    '등록함' : [50, 60],
    '등록 안함' : [30, 40]
}, index = ['빅데이터 분석기사 캠프','정보처리기사 캠프'])
df
```

로우 데이터가 주어졌을 때

```python
import pandas as pd
df = pd.DataFrame({
    '캠프' : ['빅분기'] * 80 + ['정처리'] * 100,
    '등록여부' : ['등록'] * 50 + ['등록안함']*30 + ['등록'] * 60 + ['등록안함']*40
})
df = pd.crosstab(df['캠프'],df['등록여부'])

from scipy import stats
res = stats.chi2_contingency(df)

print(res[0])
print(res[1])
print('채택')
```

#다중 성형 회귀

```python
import pandas as pd
df = pd.DataFrame({
  '할인율': [28, 24, 13, 0, 27, 30, 10, 16, 6, 5, 7, 11, 11, 30, 25, 4, 7, 24, 19, 21, 6, 10, 26, 13, 15, 6, 12, 6, 20, 21],
  '온도': [15, 34, 15, 22, 29, 30, 14, 17, 28, 29, 19, 19, 34, 10, 29, 28, 12, 25, 32, 28, 22, 16, 30, 11, 16, 18, 16, 33, 12, 22],
  '광고비' : [342, 666, 224, 764, 148, 499, 711, 596, 797, 484, 986, 347, 146, 362, 642, 591, 846, 260, 560, 941, 469, 309, 730, 305, 892, 147, 887, 526, 525, 884],
  '주문량': [635, 958, 525, 25, 607, 872, 858, 732, 1082, 863, 904, 686, 699, 615, 893, 830, 856, 679, 918, 951, 789, 583, 988, 631, 866, 549, 910, 946, 647, 943]
})
```

```python
from statsmodels.formula.api import ols
model = ols('주문량~할인율+온도+광고비', data=df).fit()
model.summary()
```

```python
#1. 할인율과 온도의 상관계수
print(round(df['할인율'].corr(df['온도']), 2))

#2. 모델의 결정계수
print(round(model.rsquared, 2))

#3. 각 변수의 회귀 계수
print(round(model.params['할인율'], 2))
print(round(model.params['온도'], 2))
print(round(model.params['광고비'], 2))

#4. 모델의 절편
print(round(model.params['Intercept'], 4))

#5. 온도의 회귀 계수 p-value
print(round(model.pvalues['온도'], 4))

#6. 할인율 10, 온도 20, 광고비 500,000 -> 예측
model.predict([10, 20, 50])
```
