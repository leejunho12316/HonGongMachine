```python
import pandas as pd
import numpy as np

# 데이터 생성 (복사 후 실행)
data1 = {
    'Group': ['A']*30 + ['B']*30 + ['C']*30,
    'Value': [98, 101, 99, 102, 100, 97, 99, 103, 101, 100, 98, 102, 101, 99, 100, 104, 102, 98, 99, 101, 100, 102, 99, 101, 100, 97, 103, 100, 101, 99,
              110, 112, 108, 115, 111, 113, 109, 114, 110, 112, 116, 111, 113, 108, 112, 110, 114, 111, 113, 109, 112, 115, 110, 113, 111, 108, 114, 110, 112, 113,
              105, 107, 106, 104, 108, 105, 107, 103, 106, 105, 109, 104, 107, 106, 105, 102, 108, 106, 104, 107, 105, 109, 103, 106, 105, 108, 104, 107, 106, 105]
}
df_prob1 = pd.DataFrame(data1)

data2 = {
    'Method': ['M1', 'M1', 'M1', 'M1', 'M1', 'M1', 'M1', 'M1', 'M1', 'M1',
               'M2', 'M2', 'M2', 'M2', 'M2', 'M2', 'M2', 'M2', 'M2', 'M2'],
    'Result': ['Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail', 'Pass', 'Pass', 'Fail', 'Pass',
               'Fail', 'Fail', 'Pass', 'Fail', 'Fail', 'Fail', 'Pass', 'Fail', 'Fail', 'Fail']
}
df_prob2 = pd.DataFrame(data2)

data3 = {
    'Class': ['X']*20 + ['Y']*20 + ['Z']*20,
    'Score': [70, 72, 68, 75, 71, 73, 69, 74, 70, 72, 68, 76, 71, 73, 69, 74, 70, 72, 68, 75,
              85, 82, 88, 85, 81, 83, 89, 84, 80, 82, 88, 86, 81, 83, 89, 84, 80, 82, 88, 85,
              60, 95, 62, 92, 61, 93, 59, 94, 60, 92, 58, 96, 61, 93, 59, 94, 60, 92, 58, 96]
}
df_prob3 = pd.DataFrame(data3)
```

문제 1. 앤더슨-달링 검정과 비모수 검정

df_prob1 데이터는 3가지 그룹(A, B, C)의 생산 수치를 나타낸다. 다음 단계에 따라 분석을 수행하고 물음에 답하시오.

Group이 'A'인 데이터에 대해 scipy.stats.anderson을 사용하여 정규성 검정을 수행하라.

검정 결과 도출된 검정 통계량(statistic)을 출력하라.

세 그룹(A, B, C) 간의 평균 차이가 있는지 검정하려 한다. 정규성 검정 결과와 상관없이, 세 집단의 분포가 정규성을 만족하지 않는다고 가정하고 **비모수 검정(Kruskal-Wallis Test)**을 수행하여 검정 통계량을 구하라.

최종 답안: A그룹의 Anderson 검정 통계량과 Kruskal-Wallis 검정 통계량을 더한 값을 소수점 둘째 자리에서 반올림하여 출력하시오.

```python
df_a = df_prob1[df_prob1['Group'] == 'A']['Value']
df_b = df_prob1[df_prob1['Group'] == 'B']['Value']
df_c = df_prob1[df_prob1['Group'] == 'C']['Value']

#1
from scipy import stats
res = stats.anderson(df_a)
print("검정통계량", res[0])

#2
#정규성 만족 안한다 가정 -> kruskal
s,p = stats.kruskal(df_a, df_b, df_c)
print(s)
print(format(p,'.10f'), '적어도 하나의 그룹의 평균은 차이가 있음')

#3
print(round(res[0] + s, 1))
```

제 2. 카이제곱 검정과 피셔의 정확 검정

df_prob2 데이터는 두 가지 교육 방법(M1, M2)에 따른 결과(Pass, Fail)를 나타낸다.

Method와 Result에 대한 분할표(Cross Table)를 생성하라.

생성된 분할표를 바탕으로 카이제곱 검정(Chi-square test)을 수행했을 때, 기대 빈도(Expected Frequency)가 5보다 작은 셀의 개수를 구하라.

위 2번의 결과에 따라 적절한 검정 방법(카이제곱 검정 또는 피셔의 정확 검정)을 선택하여 수행하고, 양측 검정(Two-sided) 기준 p-value를 도출하라.

최종 답안: 기대 빈도가 5 미만인 셀의 개수와 선택된 검정 방법의 p-value를 곱한 값을 소수점 넷째 자리에서 반올림하여 출력하시오.

```python
#1
df_c = pd.crosstab(df_prob2['Method'], df_prob2['Result'])

#2
from scipy import stats
s,p,d,exp = stats.chi2_contingency(df_c)
#print(exp)
print('2개, 전체의 50%')

#3
#기대빈도 5개미만 20%이상 : fisher 정확검정
s,p = stats.fisher_exact(df_c, alternative='two-sided')
print(s,p)

#4
print(round(2*p, 3))
```

문제 3. 등분산 검정과 Welch's ANOVA, 사후 검정

df_prob3 데이터는 세 반(X, Y, Z)의 성적 데이터를 나타낸다.

세 그룹(X, Y, Z)에 대해 stats.bartlett을 사용하여 등분산 검정을 수행하고 p-value를 구하라.

등분산 가정 충족 여부와 관계없이, 이 데이터가 등분산을 만족하지 않는다고 가정하고 Welch's ANOVA를 수행하여 p-value를 구하라. (라이브러리는 pingouin을 사용한다.)

세 그룹 간 평균 차이가 통계적으로 유의하다면, Bonferroni 방법을 사용하여 사후 검정을 수행하라. X반과 Y반 사이의 p-value(uncorrected가 아닌 보정된 값)를 구하라.

최종 답안: Bartlett 검정의 p-value, Welch's ANOVA의 p-value, 사후 검정(X-Y)의 p-value 세 값을 모두 더하여 소수점 셋째 자리에서 반올림하여 출력하시오.

```python
df_X = df_prob3[df_prob3['Class'] == 'X']['Score']
df_Y = df_prob3[df_prob3['Class'] == 'Y']['Score']
df_Z = df_prob3[df_prob3['Class'] == 'Z']['Score']

#1
from scipy import stats
s,p = stats.bartlett(df_X, df_Y, df_Z)
print(format(p,'.10f')) #등분산 아님
res1 = p

#2
import pingouin as pg
res = pg.welch_anova(data=df_prob3, dv='Score', between='Class')
print(format(res.iloc[0, -2], '.10f')) #적어도 하나 평균 차이 있음
res2 = res.iloc[0, -2]

#3
import scipy
from statsmodels.sandbox.stats.multicomp import MultiComparison
comp = MultiComparison(df_prob3['Score'], df_prob3['Class'])
res = comp.allpairtest(scipy.stats.ttest_ind, method='bonf')
res3 = pd.DataFrame(res[0].data).iloc[1,4]

print(round(res1 + res2 + res3, 2))
```

```python
import pandas as pd
import numpy as np

# 데이터 생성 (복사 후 실행)
np.random.seed(42)

# df_prob4: 광고비(Ad)와 매출(Sales) 데이터
n4 = 100
ad_spend = np.random.normal(50, 10, n4)
sales = 20 + 3 * ad_spend + np.random.normal(0, 15, n4) # 강한 양의 상관관계
df_prob4 = pd.DataFrame({'Ad': ad_spend, 'Sales': sales})

# df_prob5: 다이어트 약 복용 전(Before)과 후(After) 몸무게
n5 = 30
before = np.random.normal(80, 5, n5)
after = before - np.random.normal(2, 1, n5) # 평균적으로 2kg 감량
# 데이터에 인위적인 이상치 추가로 정규성 깨뜨리기 시도
after[0] = 50
df_prob5 = pd.DataFrame({'Before': before, 'After': after})

# df_prob6: 공부 시간(Study), 수면 시간(Sleep)에 따른 합격 여부(Pass) - 로지스틱 회귀용
n6 = 200
study = np.random.uniform(1, 10, n6)
sleep = np.random.normal(7, 1, n6)
# 합격 확률 생성 (공부 시간이 길수록, 수면 시간이 적당할수록 확률 증가)
logits = -5 + 1.2 * study + 0.5 * sleep
probs = 1 / (1 + np.exp(-logits))
pass_val = [1 if p > 0.5 else 0 for p in probs]
df_prob6 = pd.DataFrame({'Study': study, 'Sleep': sleep, 'Pass': pass_val})
```

문제 4. 상관분석과 단순선형회귀분석

df_prob4는 광고비(Ad)와 매출(Sales) 데이터를 나타낸다.

두 변수 간의 피어슨 상관계수(Pearson Correlation Coefficient)를 구하라.

Sales를 종속변수, Ad를 독립변수로 하는 단순선형회귀모델을 statsmodels를 사용하여 적합(fit)하라.

회귀분석 결과에서 **독립변수(Ad)의 회귀계수(slope)에 대한 t-검정 통계량(t-statistic)**을 구하라.

최종 답안: 상관계수와 회귀계수의 t-검정 통계량을 더한 값을 소수점 셋째 자리에서 반올림하여 출력하시오

```python
#print(df_prob4)
#1
res1 = df_prob4['Ad'].corr(df_prob4['Sales'], method='pearson')

#2
from statsmodels.formula.api import ols
model = ols('Sales~Ad',data = df_prob4).fit()
res2 = model.tvalues[1]

#최종
print(round(res1+res2, 2))
```

문제 5. 대응표본 검정 (Paired Test)

df_prob5는 동일한 집단의 다이어트 약 복용 전(Before)과 후(After)의 몸무게 변화를 나타낸다.

변화량(Before - After) 변수를 생성하라.

생성한 변화량 변수에 대해 shapiro 검정을 수행하여 정규성 여부를 판단하라. (유의수준 0.05 기준)

정규성 검정 결과에 따라 적절한 검정 방법(대응표본 t-검정 또는 윌콕슨 부호순위 검정)을 선택하여 복용 전후의 차이가 유의미한지 검정하라. (양측 검정, alternative='two-sided')

최종 답안: Shapiro-Wilk 검정의 p-value가 0.05보다 작으면 1, 크면 0을 선택하고, 이를 선택된 검정 방법의 p-value와 더한 값을 소수점 넷째 자리에서 반올림하여 출력하시오. (즉, 정규성 만족 안 함(1) + 검정 p-value 또는 정규성 만족함(0) + 검정 p-value)

```python
#1
df_prob5['difference'] = df_prob5['Before'] - df_prob5['After']

#2
from scipy import stats
s,p = stats.shapiro(df_prob5['difference'])
print(s,format(p,'.10f'))
print('정규하지 않음')

#3
# 정규X -> wilcoxon
s,p = stats.wilcoxon(df_prob5['Before'], df_prob5['After'], alternative='two-sided')
print(s,format(p,'.10f'))
print('귀무가설 기각, 차이가 유의미함.')

#최종
print(round(1 + p, 3))
```

문제 6. 로지스틱 회귀분석과 오즈비(Odds Ratio)

df_prob6은 공부 시간(Study)과 수면 시간(Sleep)이 시험 합격(Pass)에 미치는 영향을 조사한 데이터다.

Pass를 종속변수로, Study와 Sleep을 독립변수로 하는 로지스틱 회귀모형을 학습하라. (statsmodels의 logit 함수 사용)

학습된 모델에서 Study 변수의 통계적으로 유의한지 판단하기 위한 p-value를 구하라.

Study 변수의 계수(coef)를 활용하여, 공부 시간이 1단위 증가할 때 합격할 오즈(Odds)가 몇 배 증가하는지 나타내는 **오즈비(Odds Ratio)**를 구하라.

최종 답안: Study 변수의 p-value와 Study 변수의 오즈비(Odds Ratio)를 곱한 값을 소수점 셋째 자리에서 반올림하여 출력하시오.

```python
import numpy as np
#1
from statsmodels.formula.api import logit
model = logit('Pass~Study+Sleep',data=df_prob6).fit()

#2
print(model.pvalues['Study'])
pval = model.pvalues['Study']

#3
log_odds = model.params['Study']
odds = np.exp(log_odds)
print(format(odds, '.10f'))

#최종
print(round(pval*odds,2))
```
