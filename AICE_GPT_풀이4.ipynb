```python
import pandas as pd
df = pd.read_csv('auto_mpg.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#1
sns.countplot(data=df, x='origin')
plt.show()

#2
sns.histplot(data=df, x='mpg')
plt.show()
```

```python
#3
sns.boxplot(data=df, y='weight')
plt.show()
#4
sns.violinplot(data=df, y='horsepower')
```

```python
#5
df['model_year'].value_counts().sort_index().plot(kind='bar')

df_temp = df['model_year'].value_counts().sort_index()
plt.bar(x=df_temp.index, height = df_temp)
plt.show()
```

```python
#6
df_temp = df.groupby('cylinders')['mpg'].mean().reset_index()
sns.barplot(data=df_temp, x='cylinders',y='mpg')
plt.show()

#7
sns.scatterplot(data=df, x='displacement', y='mpg')
plt.show()
```

```python
#8
sns.boxplot(data=df, x='origin', y='acceleration')
plt.show()
```

```python
#9
df_temp = df.groupby('model_year')['mpg'].mean().reset_index()
sns.lineplot(data=df_temp, x='model_year', y='mpg')
```

```python
#10
sns.scatterplot(data=df, x='horsepower', y='weight', hue='origin')
```
