#Concat 연습

```python
df1 = pd.DataFrame({
    'col1' : ['사과','배','감','수박','메론'],
    'col2' : [500, 1000, 2500, 5000, 3000]
}, index = [0,1,2,3,4])
df1
```

```python
df2 = pd.DataFrame({
    'col1' : ['수박','멜론','딸기','키워','오렌지'],
    'col2' : [5000, 3000, 1000, 600, 700],
    'col3' : ['good','good','bad','bad','normal']
}, index = [0,1,2,3,4])
df2
```

```python
pd.concat([df1, df2], join = 'outer', verify_integrity = True)
```

```python
customer = pd.DataFrame({
    'customer_id' : np.arange(6),
    'name' : ['세주아니','케이틀린','아리','노틸러스','우디르','말파이트'],
    '나이' : [20, 20, 2000, 500, 600, 900]
})

orders = pd.DataFrame({
    'customer_id' : [1, 1, 2, 2, 2, 3, 3, 1, 4, 9],
    'item' : ['5.5mm 탄', '돋보기', '한복','정기', '남자', '닻', '어항','총열','닭','IPhone 14 Pro'],
    'quantity' : [100, 1, 1, 3, 3, 1, 2, 1, 20, 1]
})
```

```python
temp = pd.merge(customer, orders, on='customer_id', how='right').groupby('item').aggregate(np.sum)
temp.sort_values(by='quantity', ascending=False)
```

```python
temp2 = pd.merge(customer, orders, on='customer_id', how='inner')
temp2 = temp2.groupby(['name','item']).sum()
temp2.loc['아리']
```

```python
customer = pd.DataFrame({
    'customer_id' : np.arange(6),
    'name' : ['James','Elly','Tom','Givert','Aiden','Brody'],
    '나이' : [40,20,21,30,31,18]
})
orders = pd.DataFrame({
    'customer_id' : [1,1,2,2,2,3,3,1,4,9],
    'item' : ['마우스','충전기','이어폰','헤드셋','전자펜','키보드','전자펜','마우스','키보드','케이스'],
    'quantity' : [1,2,1,1,3,2,2,3,2,1]
})
```

```python
temp = pd.merge(customer, orders, on='customer_id', how='left')
temp.groupby(by='item').sum()['quantity']
```

```python
temp = pd.merge(customer, orders, on='customer_id', how='inner')
temp.groupby(by = ['name','item']).sum().loc['Elly']
```
