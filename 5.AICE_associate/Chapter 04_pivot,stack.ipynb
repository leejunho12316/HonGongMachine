```python
import numpy as np
import pandas as pd
```

```python
pivot_data = pd.DataFrame({
    'cust_id' : ['cust_1','cust_1','cust_1','cust_2','cust_2','cust_2','cust_3','cust_3','cust_3'],
    'prod_cd' : ['p1','p2','p3','p1','p2','p3','p1','p2','p3'],
    'grade' : ['A','A','A','A','A','A','B','B','B'],
    'purch_amt' : [30,10,0,40,15,30,0,0,10]
})
pivot_data
```

```python
pivot_data.pivot_table(index = 'prod_cd',columns='cust_id', values='grade',aggfunc=np.sum)
```
