```python
import numpy as np
import pandas as pd
```

```python
df_original = pd.read_csv('Flight Price Prediction/Clean_Dataset.csv')
df_original.drop(df_original.columns[0], axis=1, inplace = True)
df_original.head(3)
```

# 쪼개서 새로운 변수 만들기

```python
def split_flight(flight):
    manufacture, model = flight.split('-')
    return manufacture, model
```

```python
df = df_original.copy()

df['manufacture'], df['model'] = zip(*df['flight'].apply(lambda x : split_flight(x)))
```

```python
df.head(3)
```

# 합쳐서 새로운 변수 만들기

```python
df['route'] = df.apply(lambda x : x['source_city'] + '-' + x['destination_city'], axis=1)
```

```python
df['dd'] = df.apply(lambda x : str(x['duration']) + '시간' + str(x['days_left']) + '남음', axis=1)
df.head(3)
```
