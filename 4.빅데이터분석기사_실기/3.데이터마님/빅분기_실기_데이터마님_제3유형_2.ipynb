```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/scipy2.csv')
df_A = df[df['class'] == 'A']['score']
df_B = df[df['class'] == 'B']['score']

from scipy import stats
s,p = stats.levene(df_A, df_B)
print('levene : ', s,p) #귀무가설 채택, 등분산이다

s,p = stats.bartlett(df_A, df_B)
print('bartlett : ', s,p)

s,p = stats.fligner(df_A, df_B)
print('fligner : ', s,p)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df1 = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/ind1.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/ind2.csv')

print(len(df1), len(df2))

#독립 : 정규성 -> 등분산성 ->
#1. 정규성 검정 : 데이터 100개 -> shapiro O, anderson X
from scipy import stats
s1,p1 = stats.shapiro(df1) #정규성 ㅇㅇ
s2,p2 = stats.shapiro(df2) #정규성 ㅇㅇ
print('정규성 : ', s1, p1, s2, p2)

#2. 등분산성 검정 : levene,fligner, bartlett
s3,p3 = stats.levene(df1, df2)
s4,p4 = stats.fligner(df1, df2)
s5,p5 = stats.bartlett(df1, df2)
print('levene : ',s3,p3)
print('fligner : ',s4,p4)
print('bartlett : ',s5,p5)
#등분산성 만족한다고 볼 수 있다

#3. 정규성 ㅇ 등분산성 ㅇ -> 독립 t-검정
s,p = stats.ttest_ind(df1, df2, equal_var = True, alternative='two-sided')
print(s,p) #귀무 : 동일하다, 대립 : 안하다. 귀무 기각 -> 평균 동일하지 않다.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/scipy5.csv')
df_A = df['A']
df_B = df['B'].dropna()

#1. 정규성 검정
from scipy import stats #,anderson
s1,p1 = stats.shapiro(df_A)
s2,p2 = stats.shapiro(df_B)
print(s1,format(p1,'.10f'),s2,p2) #둘다 정규성 만족 X

#2. MannWhitney U
s,p = stats.mannwhitneyu(df_A, df_B, alternative='two-sided')
print(s)
print(p) #귀무가설 채택 -> 중위값은 동일하다
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/ind3.csv')

df_a = df[df['group'] == 'a']['data']
df_b = df[df['group'] == 'b']['data']

#1. 정규성 검정 shapiro O / anderson X
from scipy import stats
s1, p1 = stats.shapiro(df_a) # 정규성 ㅇ
s2, p2 = stats.shapiro(df_b) #정규성 ㅇ
print(s1, p1, s2, p2)

#2. 등분산 검정 levene, fligner, bartlett
from scipy import stats
s3, p3 = stats.levene(df_a, df_b)
s4, p4 = stats.fligner(df_a, df_b)
s5, p5 = stats.bartlett(df_a, df_b)
print(s3, p3) #등분산이 아니다
print(s4, p4) #등분산이 아니다
print(s5, p5) #등분산이 아니다

#3. Welch's T-Test
s,p = stats.ttest_ind(df_a, df_b, equal_var = False, alternative= 'two-sided')
print(s,p) #안하다
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/ind6.csv')
df_a = df['a']
df_b = df['b'].dropna()
print(len(df_a), len(df_b))

from scipy import stats
#1. 정규성 검정 shapiro O / anderson X
s1,p1 = stats.shapiro(df_a) #정규성 ㅇ
s2,p2 = stats.shapiro(df_b) #정규성 ㅇ
print(s1, p1, s2, p2)

#2. 등분산 검정
s3, p3 = stats.levene(df_a, df_b)
s4, p4 = stats.fligner(df_a, df_b)
s5, p5 = stats.bartlett(df_a, df_b)
print(s3, p3) #등분산 X
print(s4, p4) #등분산 X
print(s5, p5) #등분산 ㅇ

#-> 대체로 등분산 아님
print(df_a.shape, df_b.shape)
#3. Welch's T-test
s,p = stats.ttest_ind(df_a, df_b, equal_var = False, alternative='two-sided')
print('마지막 : ',s,p)
#평균은 같다
```
