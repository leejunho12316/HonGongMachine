```python
import torch
import numpy as np
```

#Tensor의 기초

1. 데이터 -> tensor

```python
#리스트 -> tensor
data = [[1,2],[3,4]]
x_data = torch.tensor(data)

#numpy -> tensor
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
```

```python
#속성 유지하고 데이터 생성하기
x_ones = torch.ones_like(x_data) #shape
print(x_ones)

x_rand = torch.rand_like(x_data, dtype=torch.float) #daratype, shape
print(x_rand)
```

2. 무작위 tensor

```python
#무작위 만들기
shape = (2,3,)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(rand_tensor)
print(ones_tensor)
print(zeros_tensor)
```

3. tensor 접근

```python
tensor = torch.rand(4,4)
print(tensor)
print(tensor[0,:])
print(tensor[:,0])
print(tensor[:, -1])
print(tensor[..., -1])
```

4. tensor 속성

```python
tensor = torch.rand(3,4)
print(tensor.shape, tensor.dtype, tensor.device)
```

```python
torch.cat([tensor, tensor, tensor], dim=1)
```

```python
torch.stack([tensor,tensor,tensor])
```

5. 연산

```python
y1 = tensor @ tensor.T #@ matmul : 행렬곱 T: 전치
tensor.matmul(tensor.T)


y3 = torch.rand_like(y1)
torch.matmul(tensor, tensor.T, out=y3)
```

```python
tensor * tensor
tensor.mul(tensor)
```

```python
tensor.sum().item()
```

#Numpy 변환

```python
#tensor와 numpy배열은 메모리 공간을 공유하기 때문에 tensor 데이터 numpy 변환 가능. 단, tensor 변경시 numpy 데이터도 변함.
torch_data = torch.ones(5)
numpy_data = torch_data.numpy()
print(numpy_data)

torch_data.add_(1)
print(numpy_data)
```

```python
#numpy배열을 tensor로 변경했을 때도 마찬가지고 numpy변경시 tensorflow데이터도 변함.
n = np.ones(5)
t = torch.from_numpy(n)
print(n, t)

np.add(n, 1, out=n)
print(n, t)
```
