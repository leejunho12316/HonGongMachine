```python
import os
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms

device = 'cpu'
```

```python
class NeuralNetwork(nn.Module):
  def __init__(self):
    super().__init__()
    self.flatten = nn.Flatten()
    self.linear_relu_stack = nn.Sequential(
        nn.Linear(28*28, 512),
        nn.ReLU(),
        nn.Linear(512, 512),
        nn.ReLU(),
        nn.Linear(512,10)
    )

  def forward(self, x):
    x = self.flatten(x)
    logits = self.linear_relu_stack(x)
    return logits

model = NeuralNetwork().to(device)
print(model)

X = torch.rand(1, 28, 28, device = device) #데이터
logits = model(X) #모델 통과
pred_probab = nn.Softmax(dim=1)(logits) #확률 출력
y_pred = pred_probab.argmax(1) #결과 출력

print('data : ', X[0][:2], ',,,,,,,,,,,,,,,,,,,,')
print('model result : ', logits)
print('Softmax : ', pred_probab)
print('result : ', y_pred)
```

```python
rand_image = torch.rand(3,28,28)
print('original : ', rand_image.shape)

flatten = nn.Flatten()
flat_image = flatten(rand_image)
print('flattened : ', flat_image.shape)
print(flat_image)

print()

layer1 = nn.Linear(in_features = 28*28, out_features = 20)
hidden = layer1(flat_image)
print('hidden : ', hidden.shape)
print(hidden)

print()

relu = nn.ReLU()(hidden)
print('relu : ', relu.shape)
print(relu)
```

```python
seq_modules = nn.Sequential(
    flatten,
    layer1,
    nn.ReLU(),
    nn.Linear(20,10)
)
input_image = torch.rand(3,28,28)
logits = seq_modules(input_image)
logits
```
