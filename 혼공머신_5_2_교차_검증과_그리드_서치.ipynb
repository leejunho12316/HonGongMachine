```python
import pandas as pd
wine = pd.read_csv('https://bit.ly/wine_csv_data')

wine_input = wine[['alcohol','sugar','pH']].to_numpy()
wine_target = wine['class'].to_numpy()
```

#교차검증

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
from sklearn.model_selection import StratifiedKFold
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state = 42)

train_input, test_input, train_target, test_target = train_test_split(wine_input, wine_target, random_state = 42)
res = cross_validate(dt, train_input, train_target,
                     cv = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 42))
res
```

#그리드서치

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

train_input, test_input, train_target, test_target = train_test_split(wine_input, wine_target, random_state = 42)

dt = DecisionTreeClassifier(random_state = 42)
params = {'min_impurity_decrease' : [0.0001, 0.0002, 0.0003, 0.0004, 0.0005]}

gs = GridSearchCV(dt, params, n_jobs = -1)
gs.fit(train_input, train_target)
```

```python
#가장 점수 높은 모델
dt_best = gs.best_estimator_
print(dt_best.predict(train_input[:5]))
print(dt_best.score(train_input, train_target))
```

```python
import numpy as np

#가장 점수 높은 하이퍼파라미터
print(gs.best_params_)

best_index = np.argmax(gs.cv_results_['mean_test_score'])
print(gs.cv_results_['params'][best_index])

#각 하이퍼파라미터 별 평균 점수
print(gs.cv_results_['mean_test_score'])
```

#고급 그리드서치 (하이퍼파라미터 여러개)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

train_input, test_input, train_target, test_target = train_test_split(wine_input, wine_target, test_size = 0.2, random_state = 42)

dt = DecisionTreeClassifier(random_state = 42)

params = {'min_impurity_decrease' : np.arange(0.0001, 0.001, 0.0001),
          'max_depth' : range(5,20,1),
          'min_samples_split' : range(2,100,10)
          }
gs = GridSearchCV(dt, params, n_jobs = -1)
gs.fit(train_input, train_target)


```

```python
dt_best = gs.best_estimator_
print("가장 좋은 파라미터 : ", gs.best_params_)
print("점수 : ", gs.cv_results_['mean_test_score'])
print("가장 높은 점수 : ", gs.best_score_)
```

#랜덤 서치

```python
from scipy.stats import uniform, randint

#정수, 실수 랜덤 출력
print(randint(0,10).rvs(100), "\n\n", uniform(0,1).rvs(100))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier

params = {'min_impurity_decrease' : uniform(0.0001, 0.001),
          'max_depth' : randint(20, 50),
          'min_samples_split' : randint(2,25),
          'min_samples_leaf' : randint(1,25)
          }

dt = DecisionTreeClassifier(random_state = 42, splitter = 'random')
gs = RandomizedSearchCV(dt, params, n_iter = 100, n_jobs=-1, random_state = 42)
gs.fit(train_input, train_target)
```
