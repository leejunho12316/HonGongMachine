#나의 풀이

```python
def solution(graph, start):
  queue = [start]
  visited = []

  while len(queue) != 0:
    print('before', visited)

    q = queue.pop(0)

    if q not in visited:
      for pair in graph:
        if pair[0] == q:
          queue.append(pair[1])

    visited.append(q)
    print('after', visited)
    print()


solution([(1,2),(1,3),(2,4), (2,5), (3,6), (3,7), (4,8), (5,8), (6,9), (7,9)], 1)
```

#BFS 답안

```python
from collections import defaultdict, deque

def solution(graph, start):

  #인접 리스트 만들기
  add_list = defaultdict(list)
  for u, v in graph:
    add_list[u].append(v)



  #너비 우선 탐색 함수
  def bfs(start):
    visited = set()
    visited.add(start)
    queue = deque([start])

    while queue:
      node = queue.popleft()
      for neighbor in add_list.get(node, []):
        if neighbor not in visited:
          queue.append(neighbor)
          visited.add(neighbor)

    print(visited)

  bfs(start)



solution([(1,2),(1,3),(2,4), (2,5), (3,6), (3,7), (4,8), (5,8), (6,9), (7,9)], 1)
```

### 1 DFS

```python
from collections import defaultdict

def solution(graph, start):
  #인접 그래프 전환
  add_dict = defaultdict(list)
  for u, v in graph:
    add_dict[u].append(v)


  def dfs(node, visited, result):
    visited.add(node)
    result.append(node)

    for p in add_dict.get(node, []):
      if p not in visited:
        dfs(p, visited, result)


  visited = set()
  result = []

  dfs(start, visited, result)
  print(result)


solution([['A','B'], ['B','C'], ['C','D'], ['D','E']], 'A')

```

### 2 DFS

```python
from collections import defaultdict

def solution(graph, start):
  #인접 행렬
  adj_dict = defaultdict(list)
  graph = sorted(graph, reverse = True)
  for u, v in graph:
    adj_dict[u].append(v)

  #깊이 우선 탐색
  visited = set()
  stack = []
  stack.append(start)
  result = []

  while(stack):
    q = stack.pop()
    visited.add(q)
    result.append(q)

    for res in adj_dict.get(q, []):
      if res not in visited:
        stack.append(res)
    print(stack)

  print(result)



solution([['A','B'], ['A','C'], ['B','D'], ['B','E'], ['C','F'], ['E','F']], 'A')
```
