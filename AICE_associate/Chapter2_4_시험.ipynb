#데이터 재구성하는 2가지 방법

```python
df = pd.DataFrame({
    'name' : ['이','준','호','귀','모','니','방'],
    'age' : [10,51,66,23,51,26,12],
    'level' : [1,2,1,3,1,5,2]
})
df.pivot(index = 'name', columns = 'age', values = 'level')
```

```python
df.pivot_table(index = 'name', columns = 'age', values = 'level', aggfunc = np.sum)
```

#index와 column 바꾸는 법

```python
df.set_index(['name','age']).unstack(0).stack(1)
```

#데이터 병합하는 2가지 방법

```python
concat, merge
```

```python
df1 = pd.DataFrame({
    'id' :  [1,2,3,4,5],
    '이름' : ['나','너','우리','쟤네','ㅋ'],
    'age' : [61,71,71,82,83]
})

df2 = pd.DataFrame({
    'id' : [1,1,3,4,4],
    'level' : [100,120, 130, 200, 5000],
    'playtime': [666363,777745,23048202,203580,46944]
})
```

```python
pd.concat([df1,df2], axis=1)
```

```python
pd.merge(df1, df2, how='outer') #left, right, inner, outer, cross
```
