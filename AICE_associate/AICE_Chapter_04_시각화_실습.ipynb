```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flight = pd.read_csv('Clean_Dataset.csv')
flight.head(3)
```

```python
sns.catplot(x = 'airline', y = 'price', col='class', data = flight)
```

```python
sns.lmplot(x = 'duration', y = 'price', data=flight, line_kws={'color':'red'})
```

```python
sns.countplot(x='airline', hue='class', data=flight)
```

```python
sns.jointplot(x='duration', y='price', kind = 'kde', data=flight)
```

```python
df_eco = flight[flight['class'] == 'Economy']
sns.heatmap(df_eco.select_dtypes('numbers').corr())
```
