```python
from openai import OpenAI

OPENAI_API_KEY = None
client = OpenAI(api_key = OPENAI_API_KEY)
```

#멀티턴 + 멀티쿼리

```python
def generate_multi_turn_multi_queries(previous_query: str, previous_answer: str, current_query: str) -> list:
    prompt = f"""
    You're an Internet Search Helper Agent.
    Create search keywords for an Internet Search based on current_query.
    Refer to the previous_query and previous_answer align with the user's interests.

    Each Keyword must be around 3~4 words long.
    Each keyword must focus on only one topic at a time.
    All keywords must reflect the full context of the user's interest in the conversation.
    Each tool must be highly relevant to its corresponding keyword.

    previous_query: {previous_query}
    previous_answer: {previous_answer}
    current_query: {current_query}

    OUTPUT RESTRICTIONS
    1. You MUST print only python List with string data contained.
    2. NEVER print anything else except python List.
    3. Translate keywords into Korean.

    Examples)
    Output : ['서울 여행지', '전주 여행지', '제주 여행지']

    """

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "You're an Internet Search Helper Agent. Create search keywords for an Internet Search based on current_query."},
            {"role": "user", "content": prompt}
        ],
    )

    search_queries = response.choices[0].message.content.strip()
    return eval(search_queries)

# 사용 예시
previous_query = "서울 여행지 추천"
previous_answer = "서울 여행지를 추천해드릴게요~"
current_query = "전주도 알려주고 제주 관광 상품도 알려주고 서울 맛집도 알려줘"

search_queries = generate_multi_turn_multi_queries(previous_query, previous_answer, current_query)

for idx, query in enumerate(search_queries, 1):
    print(f"검색어{idx}: {query}")
```

#Function Calling

```python
def match_queries_to_tools(tools, previous_query, previous_answer, current_query):
    tool_info = "\n".join(
        [f"- {tool['name']}: {tool['description']}" for tool in tools]
    )

    prompt = f"""
    You're an Internet Search Helper Agent.
    Create search keywords for an Internet Search based on current_query.
    Refer to the previous_query and previous_answer align with the user's interests.

    Each Keyword must be around 3~4 words long.
    Each keyword must focus on only one topic at a time.
    All keywords must reflect the full context of the user's interest in the conversation.
    Each tool must be highly relevant to its corresponding keyword.

    Tools that you can use are:
    {tool_info}

    previous_query: "{previous_query}"
    previous_answer: "{previous_answer}"
    current_query: "{current_query}"

    OUTPUT RESTRICTIONS
    1. You MUST print only 'tool name and keywords' in each line.
    2. NEVER print anything else except tools and keywords.
    3. Translate keywords into Korean.

    Examples)
    TourSpotFinder: 홍대 근처 산책하기 좋은 곳
    FoodieFinder: 홍대 근처 분위기 좋은 식당
    FoodieFinder: 홍대 근처 디저트가 맛있는 카페
    """

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "You're a Internet Search Helper Agent. You're an Internet Search Helper Agent. Create search keywords for an Internet Search based on current_query."},
            {"role": "user", "content": prompt}
        ],
    )

    matched_results = response.choices[0].message.content.strip().split("\n")
    return [line.strip() for line in matched_results if line.strip()]

# 도구 정의
tools = [
    {"name": "TourSpotFinder", "description": "관광지나 산책로 등 여행 및 휴식 장소를 추천합니다."},
    {"name": "FoodieFinder", "description": "맛집이나 카페 등 음식과 관련된 장소를 추천합니다."},
    {"name": "NoTool", "description": "매칭될 도구가 없다면 여기에다가 매칭하세요."},
]

# 예시 입력
previous_query = "서울 관광지 추천해줘"
previous_answer = "홍대 어때요?"
current_query = "그럼 근처에 산책하기 좋은 곳이랑 분위기 좋은 식당은 어디고, 거기 아니더라도 디저트가 맛있는 카페는?"

# 함수 호출
result = match_queries_to_tools(tools, previous_query, previous_answer, current_query)

# 결과 출력
for item in result:
    print(item)
```
