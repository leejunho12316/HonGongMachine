#8. 로지스틱 회귀 (Logistic Regression)

##문제 다음 데이터를 사용하여 로지스틱 회귀 분석을 수행하고 아래 질문에 답하시오. (종속변수: pass_fail, 독립변수: study_time, attendance_rate, past_score)

모든 독립변수를 사용하여 로지스틱 회귀 모델을 적합하고, past_score 변수의 회귀계수(coefficient)를 구하시오.

적합된 모델의 이탈도(deviance)를 구하시오. (힌트: 이탈도 = -2 * Log-Likelihood)

새로운 데이터 (study_time=6, attendance_rate=0.8, past_score=77)에 대해 합격(1)할 확률을 구하시오.

적합된 모델을 사용하여 전체 데이터에 대한 예측 확률을 구하고, 실제 값과 비교하여 ROC-AUC 값을 계산하시오.

```python
import pandas as pd

# 데이터 생성
data = pd.DataFrame({
    'study_time': [2, 3, 5, 1, 6, 7, 4, 5, 8, 9],
    'attendance_rate': [0.8, 0.9, 0.7, 1.0, 0.8, 0.9, 0.6, 0.7, 1.0, 0.9],
    'past_score': [60, 70, 80, 50, 85, 90, 75, 65, 95, 88],
    'pass_fail': [0, 0, 1, 0, 1, 1, 0, 1, 1, 1]
})
```

```python
from statsmodels.formula.api import logit
model = logit('pass_fail~study_time+attendance_rate+past_score', data=data).fit()
print('1. ', model.params['past_score'])
print('2. ', format(-2*model.llf, '.20f'))

new_data = pd.DataFrame({'study_time' : [6], 'attendance_rate' : [0.8], 'past_score' : [77]})
print('3. ', model.predict(new_data)[0])

data['pred'] = model.predict(data)

from sklearn.metrics import roc_auc_score
print('4. ',roc_auc_score(data['pass_fail'], data['pred']))
```

답

```python
import pandas as pd
from statsmodels.formula.api import logit
from sklearn.metrics import roc_auc_score

# 데이터는 위 코드에서 생성했다고 가정
# data = pd.DataFrame(...)

# 1. 모델 적합 및 past_score 계수
model = logit('pass_fail ~ study_time + attendance_rate + past_score', data=data).fit(disp=0) # disp=0은 최적화 메시지 끄기
print(f"1. past_score 계수: {model.params['past_score']:.4f}")

# 2. 이탈도 (deviance)
deviance = -2 * model.llf
print(f"2. 이탈도: {deviance:.4f}")

# 3. 새로운 데이터 예측
new_data = pd.DataFrame({
    'study_time': [6],
    'attendance_rate': [0.8],
    'past_score': [77]
})
pred_prob = model.predict(new_data)
print(f"3. 합격 확률: {pred_prob[0]:.4f}")

# 4. ROC-AUC
pred_probs_full = model.predict(data)
roc_auc = roc_auc_score(data['pass_fail'], pred_probs_full)
print(f"4. ROC-AUC: {roc_auc:.4f}")
```

#7-2. 다중 선형 회귀 (Multiple Linear Regression)

##문제 다음 데이터는 주택의 크기, 방 개수, 건물 연령에 따른 주택 가격을 나타낸다. OLS 회귀 분석을 수행하고 아래 질문에 답하시오. (종속변수: house_price, 독립변수: size, num_rooms, age)

모든 독립변수를 사용하여 OLS 모델을 적합하고, size 변수의 회귀계수(coefficient)를 구하시오.

적합된 모델의 잔차(residual) 분석을 수행하고, 잔차의 표준편차를 구하시오.

1번 모델에서 새로운 데이터 (size=1700, num_rooms=3, age=5)에 대해 house_price의 신뢰구간 하한(97% 신뢰수준)을 구하세요.

1번 모델에서 새로운 데이터 (size=1700, num_rooms=3, age=5)에 대해 house_price의 예측구간 상한(97% 신뢰수준)을 구하세요.

```python
import pandas as pd

# 데이터 생성
data = pd.DataFrame({
    'size': [1500, 2000, 1200, 1800, 1600, 2200, 1400, 1900, 1300, 2100],
    'num_rooms': [3, 4, 2, 4, 3, 5, 3, 4, 3, 5],
    'age': [10, 5, 20, 8, 12, 3, 15, 7, 18, 4],
    'house_price': [300, 450, 250, 400, 350, 500, 280, 420, 270, 480]
})
```

```python
from statsmodels.formula.api import ols
model = ols('house_price~size+num_rooms+age', data=data).fit()
print('1. ',model.params['size'])

print('2. ',model.resid.std())

new_data = pd.DataFrame({'size':[1700], 'num_rooms':[3], 'age':[5]})
res = model.get_prediction(new_data).summary_frame(0.03)

print('3. ', format(res['mean_ci_lower'][0],'.4f'))
print('4. ', res['obs_ci_upper'][0])
```

답

```python
import pandas as pd
from statsmodels.formula.api import ols

# (데이터 생성 코드는 위에 있다고 가정)
# data = pd.DataFrame(...)

# 1. 모델 적합 및 'size' 계수
model = ols('house_price ~ size + num_rooms + age', data=data).fit()
print(f"1. 'size' 변수의 회귀계수: {model.params['size']:.4f}")

# 2. 잔차 표준편차
print(f"2. 잔차의 표준편차: {model.resid.std():.4f}")

# 3 & 4. 신뢰구간 및 예측구간
new_data = pd.DataFrame({'size':[1700],'num_rooms':[3],'age':[5]})
# 97% 신뢰수준이므로 alpha = 1 - 0.97 = 0.03
res = model.get_prediction(new_data).summary_frame(alpha=0.03)

print(f"3. 신뢰구간 하한: {res['mean_ci_lower'][0]:.4f}")
print(f"4. 예측구간 상한: {res['obs_ci_upper'][0]:.4f}")
```

#8-2. 점추정, 구간추정

##문제 다음 데이터는 어떤 공장에서 생산되는 부품의 일일 평균 무게(g)를 20일간 측정한 샘플입니다. 이 데이터를 사용하여 아래 질문에 답하시오.

점추정: 이 공장에서 생산되는 부품의 평균 무게를 점추정하세요. (반올림하여 소수 셋째자리까지)

구간추정: 99% 신뢰수준에서 이 부품의 평균 무게에 대한 신뢰구간을 구하세요. (반올림하여 소수 셋째자리까지)

```python
import pandas as pd

# 데이터 생성
data = pd.DataFrame({
    'part_weight': [
        150.2, 149.8, 150.5, 149.9, 150.1, 150.3, 149.7, 150.0, 150.4, 149.9,
        150.1, 150.6, 149.8, 150.2, 150.3, 150.0, 149.9, 150.1, 150.4, 150.0
    ]
})
```

```python
#1
print(data['part_weight'].mean())
```

```python
#2
sample_n = len(data)
sample_m = data['part_weight'].mean()
sample_s = data['part_weight'].std()

from scipy import stats
stats.t.interval(0.99, df=sample_n-1, loc=sample_m, scale = sample_s/(sample_n**0.5))
```

```python
import pandas as pd
from scipy import stats

# (데이터 생성 코드는 위에 있다고 가정)
# data = pd.DataFrame(...)
col_name = 'part_weight'

# 1. 점추정 (샘플 평균)
sample_m = data[col_name].mean()
print(f"1. 평균 무게 (점추정): {sample_m:.3f}")

# 2. 구간추정
sample_s = data[col_name].std()
sample_n = len(data)

# 99% 신뢰수준이므로 alpha = 0.01
confidence_level = 0.99
ci = stats.t.interval(confidence_level,
                     df = sample_n - 1,
                     loc = sample_m,
                     scale = sample_s / (sample_n ** 0.5))

print(f"2. 99% 신뢰구간: ({ci[0]:.3f}, {ci[1]:.3f})")
```
