#문제1

```python
import pandas as pd
df = pd.read_csv('churn.csv')
df.head(3)
```

```python
#문제1-1
from statsmodels.formula.api import logit
model = logit('Churn~AccountWeeks+ContractRenewal+DataPlan+DataUsage+CustServCalls+DayMins+DayCalls+MonthlyCharge+OverageFee+RoamMins', data=df).fit()

over005 = []
for i in model.pvalues.index:
  if model.pvalues.loc[i]>=0.05:
    over005.append(i)

print(len(over005))

#문제1-2
model = logit('Churn~DataUsage+DayMins', data=df).fit()
model.summary()

print(round(model.params['Intercept'] + model.params['DataUsage'] + model.params['DayMins'], 3))

#문제1-3
import numpy as np
c = model.params['DataUsage']
res = np.exp(c*5)
print(round(res, 3))

```

#문제2

```python
import pandas as pd
df = pd.read_csv('piq.csv')
df.head(3)
```

```python
from statsmodels.formula.api import ols
model = ols('PIQ~Brain+Height+Weight', data=df).fit()
model.summary()
#문제2-1
print(round(model.params['Brain'], 3))

#문제2-2
print(round(model.rsquared, 2))

#문제2-3
new_data = pd.DataFrame({
    'Brain' : [90],
    'Height' : [70],
    'Weight' : [150]
})
print(round(model.predict(new_data)[0]))
```
