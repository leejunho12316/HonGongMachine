```python
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/refs/heads/master/krdatacertificate/s1.csv')
df['Resistin'] = np.log(df['Resistin'])

df_1 = df[df['Classification'] == 1]['Resistin']
df_2 = df[df['Classification'] == 2]['Resistin']

l1 = len(df_1)
l2 = len(df_2)

v1 = df_1.var()
v2 = df_2.var()

F = v2/v1
print(round(F,3))
```

```python
res = ((l1-1)*v1 + (l2-1)*v2) / (l1 + l2 - 2)
print(round(res,3))
```

```python
from scipy import stats
s,p = stats.ttest_ind(df_1, df_2, alternative='two-sided', equal_var = True) #equal_val=True면 합동 분산 추정량을 자동으로 계산해서 씀
print(round(p,3))
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/height1.csv')

print(df)
from scipy import stats
s, p = stats.shapiro(df)
print(s,p) #정규성 ㅇㅇ

s,p = stats.ttest_1samp(df['height'], 165, alternative='two-sided')
print(s,p) #165 아니다.
```
