```python
import pandas as pd
import numpy as np

df_emp = pd.DataFrame({
    '부서': ['영업', '영업', '인사', '인사', '영업', '인사'],
    '근속연수': [6, 2, 10, 3, 7, 1],
    '성과급': [500, np.nan, 400, np.nan, np.nan, 200]
})
print(df_emp)
print()

map_max = df_emp.groupby('부서')['성과급'].transform('max')
map_min = df_emp.groupby('부서')['성과급'].transform('min')

cond = df_emp['근속연수'] >= 5
df_emp['성과급'] = df_emp['성과급'].fillna(map_max.where(cond))
df_emp['성과급'] = df_emp['성과급'].fillna(map_min)
print(df_emp)
```

```python
df_shop = pd.DataFrame({
    '카테고리': ['전자', '의류', '전자', '의류', '전자', '의류'],
    '재고량': [120, 30, 50, 150, 200, 80],
    '할인율': [0.3, np.nan, 0.1, np.nan, np.nan, 0.05]
})
print(df_shop)
map_max = df_shop.groupby('카테고리')['할인율'].transform('max')
map_min = df_shop.groupby('카테고리')['할인율'].transform('min')

df_shop['할인율'] = df_shop['할인율'].fillna(map_max.where(df_shop['재고량'] >= 100))
df_shop['할인율'] = df_shop['할인율'].fillna(map_min)

print(df_shop)
```

```python
df_school = pd.DataFrame({
    '과목': ['수학', '영어', '수학', '영어', '수학', '영어'],
    '참여도': ['상', '하', '하', '상', '상', '하'],
    '점수': [90, np.nan, 60, np.nan, np.nan, 50]
})
print(df_school)

map_max = df_school.groupby('과목')['점수'].transform('max')
map_min = df_school.groupby('과목')['점수'].transform('min')

df_school['점수'] = df_school['점수'].fillna(map_max.where(df_school['참여도'] == '상'))
df_school['점수'] = df_school['점수'].fillna(map_min)
print(df_school)
```

```python
df_game = pd.DataFrame({
    '직업': ['전사', '마법사', '전사', '마법사', '전사', '마법사'],
    '레벨': [60, 30, 45, 55, 70, 20],
    '공격력': [300, np.nan, np.nan, 250, np.nan, 100]
})
print(df_game)
print()

map_max = df_game.groupby('직업')['공격력'].transform('max')
map_min = df_game.groupby('직업')['공격력'].transform('min')
df_game['max'] = map_max
df_game['min'] = map_min

cond = df_game['레벨'] >= 50
df_game['공격력'] = df_game['공격력'].fillna(map_max.where(cond))
df_game['공격력'] = df_game['공격력'].fillna(map_min)

print(df_game)

```

```python
df_estate = pd.DataFrame({
    '동': ['강남', '서초', '강남', '서초', '강남', '서초'],
    '건축년도': [2015, 2005, 2000, 2020, 2018, 2008],
    '거래가': [20, np.nan, 10, np.nan, np.nan, 12]
})

map_max = df_estate.groupby('동')['거래가'].transform('max')
map_min = df_estate.groupby('동')['거래가'].transform('min')

cond = df_estate['건축년도'] >= 2010
df_estate['거래가'] = df_estate['거래가'].fillna(map_max.where(cond))
df_estate['거래가'] = df_estate['거래가'].fillna(map_min)

print(df_estate)
```
