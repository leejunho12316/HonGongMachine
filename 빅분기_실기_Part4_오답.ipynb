#제6회 기출문제

##문제1

```python
import pandas as pd
df = pd.DataFrame({
    '항암약':[4,4,3,4,1,4,1,4,1,4,4,2,1,4,2,3,2,4,4,4]
})
print(df.head(3))
```

```python
#문제 1-1
ratio = (df['항암약']==4).sum()/len(df)
print(ratio)

#문제 1-2
from scipy import stats
#감기약, 알려진 발생비율, expected
prob = [0.1, 0.05, 0.15, 0.7]
expected = [x*len(df) for x in prob]

#항암약, 관찰된 부작용, observed
observed = df.value_counts().sort_index().tolist()

s,p = stats.chisquare(observed, expected)
print(s)

#문제 1-3
print(p)
```

##문제2

```python
import pandas as pd
df = pd.read_csv('data6-3-2.csv')
print(df.head(3))
```

```python
from statsmodels.formula.api import ols
model = ols('temperature~solar+wind+o3', data=df).fit()
model.summary()
```

```python
#문제2-1
print(model.params['o3'])
#문제2-2
print(model.pvalues['wind'])
#문제2-3
new_data = pd.DataFrame({
    'solar' : [100],
    'wind' : [5],
    'o3' : [30]
})
res = model.predict(new_data)[0]
print(res)
```
