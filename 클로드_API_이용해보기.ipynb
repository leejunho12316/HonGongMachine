```python
import anthropic
import os
os.environ['ANTHROPIC_API_KEY'] = None
client = anthropic.Anthropic()
```

#Claude 이미지 인식

```python
import base64

def encode_image(image_path):
  with open(image_path, 'rb') as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')
```

```python
base64_image = encode_image('/content/Gemini_Generated_Image_7gynxt7gynxt7gyn.png')
```

```python
page = client.beta.models.list()
for p in page.data:
  print(p.id)
```

```python
message = client.messages.create(
    model = 'claude-sonnet-4-6',
    max_tokens=4096,
    messages=[
        {
            'role' : 'user',
            'content' : [
                {
                    'type' : 'image',
                    'source' : {
                        'type' : 'base64',
                        'media_type' : 'image/png',
                        'data' : base64_image
                    }
                },
                {
                    'type' : 'text',
                    'text' : '''
                    본인은 유튜브 릴스를 제작하는 제작자입니다.
                    당신은 유튜브 릴스 영상을 만들어야 합니다.
                    이 이미지를 주제로 하는 릴스 타임라인과 스크립트를 작성하세요.
                    다음의 규칙과 예시를 준수하세요.


                    - 규칙
                    1. 20초 길이의 릴스 스크립트일 것.
                    2. 스크립트를 적기 전 전체 내용이 어떤 주제이고 어떤 내용을 전하고자 하는지 작성하세요.
                    3. 타임라인과 스크립트 이외에 어떤 것도 출력하지 마세요.
                    4. 한 장면 당 2-3문장씩 스크립트 양을 많게 할 것.
                    '''
                }
            ]
        }
    ]
)
```

```python
context = message.content[0].text
print(context)
```
