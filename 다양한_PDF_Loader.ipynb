```python
import urllib.request
urllib.request.urlretrieve('https://github.com/chatgpt-kr/openai-api-tutorial/raw/main/ch06/2023_%EB%B6%81%ED%95%9C%EC%9D%B8%EA%B6%8C%EB%B3%B4%EA%B3%A0%EC%84%9C.pdf',
                           filename = '2023_북한인권보고서.pdf')
```

```python
from langchain.document_loaders import PyPDFLoader
loader = PyPDFLoader('2023_북한인권보고서.pdf')
pages = loader.load_and_split()
```

```python
print(len(pages))
```

```python
from langchain_community.document_loaders import PyMuPDFLoader
loader = PyMuPDFLoader('https://github.com/chatgpt-kr/openai-api-tutorial/raw/main/ch06/2023_%EB%B6%81%ED%95%9C%EC%9D%B8%EA%B6%8C%EB%B3%B4%EA%B3%A0%EC%84%9C.pdf')
pages = loader.load_and_split()
```
